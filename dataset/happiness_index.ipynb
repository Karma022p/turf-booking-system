```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
happiness_dataset = pd.read_csv('happiness_index_dataset (1).csv')
```

```python
happiness_dataset.head()
```

```python
happiness_dataset.tail()
```

```python
happiness_dataset.dropna() 
```

```python
happiness_dataset['Result'].value_counts()
happiness_dataset.groupby('Result').mean()
```

```python
X = happiness_dataset.drop(columns = 'Result', axis=1)
Y = happiness_dataset['Result']
```

```python
scaler = StandardScaler()
scaler.fit(X)
standardized_data = scaler.transform(X)
```

```python
X = standardized_data
Y = happiness_dataset['Result']
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=24)
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the support vector Machine Classifier
classifier.fit(X_train, Y_train)
```

```python
# # accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score of the test data : ', test_data_accuracy)
```
